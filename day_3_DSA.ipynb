```python
#print all thefactor of number
n=int(input("enter a number"))
lst=[]
for i in range(1,n+1):
  if n%i==0:
    lst.append(i)

print(lst)
```

```python
#frequency map
nums=[5,9,9,1,9,2,7,2,7]
dct={}
for n in nums:
  if n not in dct:
    dct[n]=1
  else:
    dct[n]+=1
print(dct)
```

```python
#number is palindrome or not
num=input("enter a string")
if num==num[::-1]:
  print("string is palindrome")
else:
  print("string is not palindrome")
```

```python
def fibonnaci(n):
  if n==0:
    return 0
  elif n==1:
      return 1
  else:
    return fibonnaci(n-2)+fibonnaci(n-1)
print(fibonnaci(5))
```

```python
class Solution:
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1

        while mid <= high:

            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1

            elif arr[mid] == 1:
                mid += 1

            else:  # arr[mid] == 2
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr
```

```python
class Solution:
    def minJumps(self, arr):
        n = len(arr)

        # Already at the last index
        if n == 1:
            return 0

        # Cannot make the first jump
        if arr[0] == 0:
            return -1

        max_reach = arr[0]
        steps = arr[0]
        jumps = 1

        for i in range(1, n):

            # If we have reached the last index
            if i == n - 1:
                return jumps

            # Update the farthest position reachable
            max_reach = max(max_reach, i + arr[i])

            # Use one step from the current jump
            steps -= 1

            # Current jump is exhausted
            if steps == 0:
                jumps += 1

                # Cannot move further
                if i >= max_reach:
                    return -1

                # Start a new jump
                steps = max_reach - i

        return -1
```

```python
#target of array
class Solution:
    def search(self, arr, x):
        # code here
        for index in range(len(arr)):
            if arr[index] == x:
                return index

        return -1
```

```python
#prime check
class Solution:
    def isPrime(self, n):
        # code here
        if n < 2:
            return False

        for i in range(2, n // 2 + 1):
            if n % i == 0:
                return False

        return True
```

```python
#number palindrome
class Solution:
    def isPalindrome(self, n):
        return str(abs(n)) == str(abs(n))[::-1]
```

```python
class Solution:
    def reverseWords(self, s):
        words = s.split(".")

        # Remove empty strings caused by extra dots
        words = [word for word in words if word != ""]

        # Reverse and join with a single dot
        return ".".join(words[::-1])
```

```python
class Solution:
    def findFloor(self, arr, x):
        arq = []

        for i in range(len(arr)):
            if arr[i] <= x:
                arq.append(arr[i])

        arr_new = sorted(arq)

        if len(arr_new) == 0:
            return -1
        else:
            return len(arr_new) - 1
```
